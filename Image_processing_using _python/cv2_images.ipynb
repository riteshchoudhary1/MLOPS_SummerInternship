```python
import numpy as np
import pandas as pd
import random as rm
import cv2
from google.colab.patches import cv2_imshow
import math
```

```python
photo = np.zeros((500,500,3))
def drawFlower(newphoto):
  x=250
  y=250
  newphoto = cv2.circle(newphoto, (x+0, y+100), 100, (255, 0, 0), 3)
  newphoto = cv2.circle(newphoto, (x+100, y+0), 100, (255, 0, 0), 3)
  newphoto = cv2.circle(newphoto, (x-100, y+0), 100, (255, 0, 0), 3)
  newphoto = cv2.circle(newphoto, (x+0, y-100), 100, (255, 0, 0), 3)
  r=int(100/math.sqrt(2))
  newphoto = cv2.circle(newphoto, (x+r, y+r), 100, (255, 0, 0), 3)
  newphoto = cv2.circle(newphoto, (x+r, y-r), 100, (255, 0, 0), 3)
  newphoto = cv2.circle(newphoto, (x-r, y-r), 100, (255, 0, 0), 3)
  newphoto = cv2.circle(newphoto, (x-r, y+r), 100, (255, 0, 0), 3)   
  return photo
```

```python
ph=drawFlower(photo)
```

```python
cv2_imshow(ph)
```
